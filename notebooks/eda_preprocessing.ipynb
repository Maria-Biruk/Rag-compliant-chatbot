```python
target_products = [
    "Credit card",
    "Checking or savings account",
    "Money transfers",
    "Money transfer, virtual currency, or money service",
    "Consumer Loan",
    "Payday loan, title loan, or personal loan",
    "Payday loan, title loan, personal loan, or advance loan"
]

filtered_df = df[df["Product"].isin(target_products)]

print(filtered_df.shape)
```

```python
filtered_df = filtered_df.dropna(
    subset=["Consumer complaint narrative"]
)

print(filtered_df.shape)
```

```python
import re
```

```python
def clean_text(text):

    # lowercase
    text = text.lower()

    # remove boilerplate phrases
    boilerplate = [
        "i am writing to file a complaint",
        "i would like to file a complaint",
        "i am submitting this complaint"
    ]

    for phrase in boilerplate:
        text = text.replace(phrase, "")

    # remove special characters
    text = re.sub(
        r"[^a-zA-Z0-9\s]",
        "",
        text
    )

    # remove extra spaces
    text = re.sub(
        r"\s+",
        " ",
        text
    )

    return text.strip()
```

```python
filtered_df["clean_narrative"] = (
    filtered_df["Consumer complaint narrative"]
    .apply(clean_text)
)
```

```python
def clean_text(text):

    # convert to string
    text = str(text)

    # remove byte markers
    text = text.replace("b'", "")
    text = text.replace("'", "")

    # lowercase
    text = text.lower()

    # remove boilerplate phrases
    boilerplate = [
        "i am writing to file a complaint",
        "i would like to file a complaint",
        "i am submitting this complaint"
    ]

    for phrase in boilerplate:
        text = text.replace(phrase, "")

    # remove special characters
    text = re.sub(
        r"[^a-zA-Z0-9\s]",
        "",
        text
    )

    # normalize spaces
    text = re.sub(
        r"\s+",
        " ",
        text
    )

    return text.strip()
```

```python
filtered_df["clean_narrative"] = (
    filtered_df["Consumer complaint narrative"]
    .apply(clean_text)
)
```
